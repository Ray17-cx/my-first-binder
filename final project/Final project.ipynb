We load the Arabica and Robusta datasets from the GitHub mirror of the CQI Coffee Quality Database.
Column names are standardized using clean_names().

```r
arabica_raw <- read.csv("arabica_ratings_raw.csv",header=TRUE) %>%
  clean_names()

robusta_raw <- read.csv("robusta_ratings_raw.csv",header=TRUE) %>%
  clean_names()
#show the 10 row
head(arabica_raw, 10)
head(robusta_raw, 10)
```

We identify columns common to both datasets and row-bind them into a unified dataset, adding a species column to preserve species identity.

```r
#combine by same variable
#make sure the same type
arabica_raw$harvest_year <- as.character(arabica_raw$harvest_year)
robusta_raw$harvest_year <- as.character(robusta_raw$harvest_year)

common_cols <- intersect(colnames(arabica_raw), colnames(robusta_raw))

coffee_raw <- bind_rows(
  arabica_raw[, common_cols] %>% mutate(species = "Arabica"),
  robusta_raw[, common_cols] %>% mutate(species = "Robusta")
)
```

We selected a subset of variables that directly support the research questions and analytical methods required in this project:

Flavor attributes (flavor, aftertaste, balance, clean_cup, cupper_points)
These variables are numeric and suitable for PCA and K-means clustering.

Total Cup Points (total_cup_points)
This serves as the response variable in the regression model.

Processing Method (processing_method)
Used to evaluate differences across washed, natural, and honey coffees (RQ2).

Country of Origin & Region (country_of_origin, region)
Used to investigate geographical differences (RQ3).

Altitude (altitude)
Included as a potential predictor of flavor and quality.

We removed certificate URLs, IDs, owner names, and other administrative or non-informative fields because they do not contribute to PCA, clustering, or regression.

```r
#select the variable
coffee_selected <- coffee_raw %>%
  select(
    species,
    country_of_origin,
    region,
    altitude,
    processing_method,
    flavor, 
    aftertaste,
    balance,
    clean_cup,
    cupper_points,
    total_cup_points
  )
coffee_selected$total_cup_points <- parse_number(coffee_selected$total_cup_points)

```

```r
#handle the missing values
coffee_selected <- coffee_selected %>%
  mutate(across(where(is.character), ~ na_if(str_trim(.), "")))
colSums(is.na(coffee_selected))
coffee_clean <- coffee_selected
```

We checked missing values using colSums(is.na(coffee_selected)). All of the flavor variables (flavor, aftertaste, balance, clean_cup, cupper_points) and our main outcome variable (total_cup_points) had no missing values. Since these variables are required for PCA, clustering, and regression, it was important that they were complete.

Some other variables, such as region, altitude, and processing_method, did have missing values. However, because these variables are not essential for the core analyses, removing rows based on them would unnecessarily reduce the dataset. For this reason, we kept all observations and treated the dataset as clean for analysis.

We saved the final cleaned dataset as coffee_clean.csv.

## PCA

```r
#PCA
coffee_pca <- coffee_clean %>%
  mutate(across(
    c(flavor, aftertaste, balance, clean_cup, cupper_points, total_cup_points),
    ~ifelse(.x == 0, NA, as.numeric(.x))
  )) %>%
  drop_na(flavor, aftertaste, balance, clean_cup, cupper_points, total_cup_points)
flavor_data <- coffee_pca %>%
  select(flavor, aftertaste, balance, clean_cup, cupper_points, total_cup_points)

```

```r
flavor_scaled <- scale(flavor_data)
pca_result <- prcomp(flavor_scaled, center = TRUE, scale. = TRUE)

summary(pca_result)

```

```r
#Scree Plot

pca_var <- pca_result$sdev^2
pca_var_exp <- pca_var / sum(pca_var)

scree_df <- data.frame(
  PC = factor(paste0("PC", 1:length(pca_var_exp)),
              levels = paste0("PC", 1:length(pca_var_exp))),
  variance_explained = pca_var_exp
)

ggplot(scree_df, aes(x = PC, y = variance_explained)) +
  geom_col(fill = "steelblue") +
  geom_point(size = 3) +
  geom_line(group = 1) +
  labs(
    title = "Scree Plot for PCA (After Removing Invalid Zero Scores)",
    x = "Principal Component",
    y = "Proportion of Variance Explained"
  ) +
  theme_minimal()

```

```r
#PCA Scatter Plot

pca_scores <- as.data.frame(pca_result$x) %>%
  bind_cols(coffee_pca %>% select(processing_method))

ggplot(pca_scores, aes(x = PC1, y = PC2, color = processing_method)) +
  geom_point(alpha = 0.7) +
  labs(
    title = "PCA Scatter Plot by Processing Method (Cleaned Correctly)",
    x = "PC1",
    y = "PC2"
  ) +
  theme_minimal()


```

### PCA Interpretation

In this project, PCA was applied to six cupping attributes:
flavor, aftertaste, balance, clean_cup, cupper_points, total_cup_points.
These variables represent standardized sensory scores assigned by certified Q graders.

After cleaning the data (removing 3 rows with impossible values such as clean_cup = 0), PCA was performed on the remaining 1337 observations.

The scree plot shows that:

PC1 explains about 70% of the total variance,

PC2 explains around 15%,

Remaining components each contribute less than 6%.

This means the majority of variation in these flavor-related attributes can be summarized by a single underlying dimension, which is expected because all six sensory scores are highly correlated and tend to move together.

Why the PCA scatter plot forms a “line” instead of a cloud

The PCA scatter plot of PC1 vs PC2 shows points arranged along a narrow band rather than a circular cloud.
This pattern is not an error; it comes from the nature of the cupping data:

All six attributes are scored using discrete increments (~0.08–0.17) rather than continuous measurements.
This creates a limited number of possible value combinations.

clean_cup only has 11 unique values, which compresses variation along one main direction.

Because all flavor attributes are positively correlated and have similar scoring scales,
most variation is shared, causing PCA to collapse onto a dominant axis (PC1).

This is typical for sensory evaluation datasets where scores represent ordinal grading levels rather than continuous physical measurements.

Interpretation of PC1 and PC2

PC1 mainly reflects overall cup quality (higher PC1 = higher scores across all flavor attributes).

PC2 captures small differences in how individual attributes vary relative to each other, but explains much less variance.

Thus, PC1 can be interpreted as a “general quality factor,” which aligns with how Q graders evaluate coffee.

## Clustering

```r
# Select the 6 flavor attributes for clustering
cluster_data <- flavor_data   
cluster_scaled <- scale(cluster_data)

```

```r
set.seed(123)

wss <- map_dbl(1:10, function(k){
  kmeans(cluster_scaled, centers = k, nstart = 20)$tot.withinss
})

elbow_df <- tibble(k = 1:10, wss = wss)

ggplot(elbow_df, aes(k, wss)) +
  geom_line() +
  geom_point(size = 2) +
  labs(
    title = "Elbow Method for Choosing k",
    x = "Number of Clusters",
    y = "Within-Cluster Sum of Squares"
  ) +
  theme_minimal()

```

```r
set.seed(123)
k3 <- kmeans(cluster_scaled, centers = 3, nstart = 20)

coffee_clustered <- coffee_pca %>%
  mutate(cluster = factor(k3$cluster))
```

```r
# Reuse PCA scores
pca_scores <- as.data.frame(pca_result$x[, 1:2]) %>%
  rename(PC1 = PC1, PC2 = PC2) %>%
  mutate(cluster = coffee_clustered$cluster)

ggplot(pca_scores, aes(x = PC1, y = PC2, color = cluster)) +
  geom_point(alpha = 0.7) +
  labs(
    title = "K-means Clusters Visualized on PCA Plot",
    x = "PC1",
    y = "PC2",
    color = "Cluster"
  ) +
  theme_minimal()
```

### Interpretation
The PCA scatter plot displays the three K-means clusters projected onto the first two principal components.
Most of the separation occurs along the PC1 axis, which aligns with our earlier PCA finding that PC1 represents overall coffee quality.

Cluster 1 and Cluster 3 occupy the higher end of PC1, corresponding to coffees with higher feature scores, while Cluster 2 is positioned toward the lower end, representing lower-quality samples.

The elongated shape of the clusters along a single direction reflects the discrete scoring system and high correlations among cupping attributes.
This confirms that the clustering structure is primarily quality-driven rather than driven by specific flavor differences or processing methods.

```r
cluster_summary <- coffee_clustered %>%
  group_by(cluster) %>%
  summarize(
    n = n(),
    mean_flavor = mean(flavor),
    mean_aftertaste = mean(aftertaste),
    mean_balance = mean(balance),
    mean_clean_cup = mean(clean_cup),
    mean_cupper_points = mean(cupper_points),
    mean_total_cup_points = mean(total_cup_points)
  )

cluster_summary
```

```r
ggplot(cluster_summary, aes(x = cluster, y = mean_total_cup_points, fill = cluster)) +
  geom_col() +
  labs(
    title = "Average Total Cup Points by Cluster",
    x = "Cluster",
    y = "Mean Total Cup Points"
  ) +
  theme_minimal()
```

### Interpretation
The bar plot shows clear differences in average total cup points across the three clusters identified by K-means.
Cluster 1 has the highest mean score, Cluster 3 follows closely behind, and Cluster 2 has the lowest average score.

This pattern indicates that the K-means algorithm effectively separated the coffee samples into groups based on overall quality.
Cluster 1 represents higher-quality coffees, Cluster 3 contains mid–high quality samples, and Cluster 2 represents coffees with lower total cup points.

Because the six cupping attributes are strongly correlated, the clusters mainly reflect different ranges of overall cup quality rather than differences in individual flavor components.

## ANOVA 

```r
anova_cluster <- aov(total_cup_points ~ cluster, data = coffee_clustered)
summary(anova_cluster)
```

### Interpretation of ANOVA for Cluster Differences
The one-way ANOVA tested whether the K-means clusters differed in overall coffee quality (total cup points). The results showed a very large F statistic (F(2, 1334) = 1191, p < 2e-16), indicating strong and statistically significant differences among the three clusters.

This means that the clusters identified by K-means represent meaningfully different groups of coffee samples. Some clusters contain coffees with much higher total cup points, while others contain lower-scoring samples.

Therefore, RQ1 is supported:
Yes, coffee samples can be grouped into meaningful flavor-based clusters.
The clusters are not arbitrary — they differ significantly in overall coffee quality.

```r
anova_flavor     <- aov(flavor ~ processing_method, data = coffee_pca)
anova_aftertaste <- aov(aftertaste ~ processing_method, data = coffee_pca)
anova_balance    <- aov(balance ~ processing_method, data = coffee_pca)
anova_total <- aov(total_cup_points ~ processing_method, data = coffee_pca)


summary(anova_total)
summary(anova_flavor)
summary(anova_aftertaste)
summary(anova_balance)
```

### Interpretation

#### 1. Total Cup Points
Result:
Pr(>F) = 0.102  not significant

There is no statistically significant difference in overall coffee quality (total cup points) across different processing methods. Although the mean scores vary slightly, processing method does not reliably predict total cup quality.

#### 2. Flavor
Result:
Pr(>F) = 0.000834  highly significant

Processing methods significantly affect flavor scores. This means washed, natural, honey, and other methods lead to systematically different flavor profiles, even though overall quality may not differ.

#### 3. Aftertaste
Result:
Pr(>F) = 0.00123  significant

Aftertaste scores vary significantly by processing method. Certain methods consistently produce coffees with stronger or weaker aftertaste.

#### 4. Balance
Result:
Pr(>F) = 0.000542  highly significant

Processing method has a strong effect on balance scores, indicating that the sensory harmony of the cup differs across processing techniques.



Across processing methods, there are significant differences in flavor, aftertaste, and balance, indicating that processing strongly shapes the sensory profile of coffee. However, total cup points do not differ significantly, meaning that while the flavor characteristics change, the overall quality level remains comparable.
